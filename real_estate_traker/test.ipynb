```python
# EUC-KR에서 UTF-8로 변환
input_file = 'original.txt'  # EUC-KR로 저장된 파일 경로
output_file = 'code.txt'  # UTF-8로 저장할 파일 경로

# 파일 변환
with open(input_file, 'r', encoding='euc-kr') as infile:
    content = infile.read()

with open(output_file, 'w', encoding='utf-8') as outfile:
    outfile.write(content)

print(f"{input_file} 파일이 UTF-8로 변환되어 {output_file}에 저장되었습니다.")
```
